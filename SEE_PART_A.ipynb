```python
class Time:
    def __init__(self,hour,minute,sec):
        self.hour=hour
        self.minute=minute
        self.sec=sec

    def __add__(self,other):
        total_sec=self.sec+other.sec+60*(self.minute+other.minute)+3600*(self.hour+other.hour)
        new_hour,rem=divmod(total_sec,3600)
        new_min,new_sec=divmod(rem,60)
        return Time(new_hour,new_min,new_sec)

    def __sub__(self,other):
        total_sec=self.sec-other.sec+60*(self.minute-other.minute)+3600*(self.hour-other.hour)
        new_hour,rem=divmod(total_sec,3600)
        new_min,new_sec=divmod(rem,60)
        return Time(new_hour,new_min,new_sec)

    def display(self):
        print(f"{self.hour:02d}:{self.minute:02d}:{self.sec:02d}")

time1=Time(5,50,15)
time2=Time(2,15,20)

print("Time 1:")
time1.display()

print("\nTime 2:")
time2.display()

sum_time=time1+time2
print("\nSum of Time1 and Time2:")
sum_time.display()

diff_time=time1-time2
print("\nDifference of Time1 and Time2:")
diff_time.display()
```

```python
class Time:
    def __init__(self, hour, min, sec):
        self.hour = hour
        self.min = min
        self.sec = sec

    def add(self, other):
        new_t = (self.sec + other.sec) + 60 * (self.min + other.min) + 3600 * (self.hour + other.hour)
        new_hr, remainder = divmod(new_t, 3600)
        new_min, new_sec = divmod(remainder, 60)

        return Time(new_hr, new_min, new_sec)

    def sub(self, other):
        new_t = (self.sec - other.sec) + 60 * (self.min - other.min) + 3600 * (self.hour - other.hour)
        new_hr, remainder = divmod(new_t, 3600)
        new_min, new_sec = divmod(remainder, 60)

        return Time(new_hr, new_min, new_sec)

    def display(self):
        print(f"hr={self.hour} min={self.min} sec={self.sec}")

# Create an instance of the Time class
Time2 = Time(2, 15, 20)
Time2.display()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read data from the result file (replace 'result.csv' with your actual file path)
result_data = pd.read_csv("results.csv")

# Display the top 10 rows of the dataset
print("Top 10 rows of the dataset:")
print(result_data.head(10))

# Count pass and fail for each subject and overall result
subjects = result_data.columns[1:-2]  # Assuming columns from 2nd to second-to-last are subjects
pass_fail_counts = {"Overall": {"Pass": 0, "Fail": 0}}

for subject in subjects:
    pass_fail_counts[subject] = {
        "Pass": (result_data[subject] >= 40).sum(),
        "Fail": (result_data[subject] < 40).sum(),
    }

    # Update overall pass/fail counts
    pass_fail_counts["Overall"]["Pass"] += pass_fail_counts[subject]["Pass"]
    pass_fail_counts["Overall"]["Fail"] += pass_fail_counts[subject]["Fail"]

# Display pass/fail counts
print("\nPass/Fail Counts:")
print(pd.DataFrame(pass_fail_counts))

# Visualize the data

# Scatter Plot of Subject1 vs Subject2
sns.scatterplot(x="English", y="Science", data=result_data)
plt.title("Scatter Plot of Subject1 vs Subject2")
plt.show()

# Line Chart of Subject-wise Pass/Fail Counts
pass_fail_df = pd.DataFrame(pass_fail_counts).T
pass_fail_df.plot(kind="bar", stacked=True)
plt.title("Pass/Fail Counts by Subject")
plt.xlabel("Subject")
plt.ylabel("Count")
plt.show()

# Bar Chart of Overall Pass/Fail Counts
overall_pass_fail = pd.DataFrame(pass_fail_counts["Overall"], index=["Overall"])
overall_pass_fail.plot(kind="bar", stacked=True)
plt.title("Overall Pass/Fail Counts")
plt.xlabel("Overall")
plt.ylabel("Count")
plt.show()

# Histogram of Total Marks
plt.hist(result_data["Total"], bins=20, color="skyblue", edgecolor="black")
plt.title("Histogram of Total Marks")
plt.xlabel("Total Marks")
plt.ylabel("Frequency")
plt.show()
```
